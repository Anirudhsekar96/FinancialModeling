```python
import pandas as pd
# read the sales data 
data = pd.read_csv('/Users/akanksha_rajput/Documents/sem2/Fin_model/sales.csv')


col = ["year", "month"]
data.drop(col, inplace = True, axis = 1)
data.head()
```

```python
# Change sales data to tidy data
df = pd.melt(data, id_vars=['date'], var_name = "Manufaturer", value_name = "Sales" )
df.head()
```

```python
# creating column for month , year and quarter

df["Year"] = df['date'].apply(lambda x : x.split(" ")[1])
df["Month"] = df['date'].apply(lambda x : x.split(" ")[0])

q1 = ["Jan", "Feb", "Mar"]
q2 = ["Apr", "May", "Jun"]
q3 = ["Jul", "Aug", "Sep"]
q4 = ["Oct", "Nov", "Dec"]

df["Quarter"] = ""
df.loc[df["Month"].isin(q1),["Quarter"]] = "Q1"
df.loc[df["Month"].isin(q2),["Quarter"]] = "Q2"
df.loc[df["Month"].isin(q3),["Quarter"]] = "Q3"
df.loc[df["Month"].isin(q4),["Quarter"]] = "Q4"



df.drop(["Month", "date"], inplace = True, axis = 1)
df.head()
```

```python
# Change Manufacturers according to other data

df_n = df.groupby("Manufaturer")

#m = list(df_m.groups.get_keys())
doc_id = list(df_n.groups.keys())
doc_id
df.loc[df["Manufaturer"]== "Audi",["Manufaturer"]] = "AUDI AG"
df.loc[df["Manufaturer"]== "BMW",["Manufaturer"]] = "Bayerische Motoren Werke AG"
df.loc[df["Manufaturer"]== "FCA",["Manufaturer"]] = "Fiat Chrysler Automobiles NV"
df.loc[df["Manufaturer"]== "Honda Group",["Manufaturer"]] = "Honda Motor Co Ltd"
df.loc[df["Manufaturer"]== "Hyundai",["Manufaturer"]] = "Hyundai Motor Co"
df.loc[df["Manufaturer"]== "Kia",["Manufaturer"]] = "Kia Motors Corp"
df.loc[df["Manufaturer"]== "Mazda",["Manufaturer"]] = "Mazda Motor Corp"
df.loc[df["Manufaturer"]== "Mitsubishi",["Manufaturer"]] = "Mitsubishi Motors Corp"
df.loc[df["Manufaturer"]== "Nissan",["Manufaturer"]] = "Nissan Motor Co Ltd"
df.loc[df["Manufaturer"]== "Subaru",["Manufaturer"]] = "Subaru Corp"
df.loc[df["Manufaturer"]== "Tesla",["Manufaturer"]] = "Tesla Inc"
df.loc[df["Manufaturer"]== "Toyota Group",["Manufaturer"]] = "Toyota Motor Corp"
df.loc[df["Manufaturer"]== "Volkswagen",["Manufaturer"]] = "Volkswagen AG"

df.head()
```

```python
# read index data and consider only revelant rows and columns

data_in = pd.read_excel('/Users/akanksha_rajput/Documents/sem2/Fin_model/index.xlsx')

df_i = data_in[1:40]
col_i = ["Ticker","Field ID", "Field Mnemonic", "Data State"]
df_i.drop (col_i, inplace = True, axis = 1)
#df_i.head()

```

```python
# preprocess index data
df_index = pd.melt(df_i, id_vars = "Description", var_name = "Year_q", value_name = "Index")
df_index.head()
df_index["Year"] = df_index['Year_q'].apply(lambda x : x.split(" ")[0])
df_index["Quarter"] = df_index['Year_q'].apply(lambda x : x.split(" ")[1])
df_index.drop("Year_q", inplace = True, axis = 1)
df_index_1 = df_index.rename(columns={"Description": "Manufaturer"})
df_index_1.head()
```

```python
# read ratio data and consider only revelant rows and columns
df_ratio = pd.read_excel('/Users/akanksha_rajput/Documents/sem2/Fin_model/ratio.xlsx')

df_ratio.drop (col_i, inplace = True, axis = 1)
df_ratio = data_ratio[2:41]
```
