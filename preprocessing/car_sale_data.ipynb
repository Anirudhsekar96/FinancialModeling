```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('Cars_sales_per_month.xlsx')
```

```python
df = df.iloc[6::,:]
df = df.applymap(lambda x: 0 if x == '--' else x)
```

```python
df.head()
```

```python
df.tail()
```

```python
option = df.iloc[0, ::].unique()[1::]
colnames = df.iloc[0, 1:]
```

```python
len(option)
```

```python
option_dict = {i: [j for j in range(116) if not df.iloc[0,j]!=i]  for i in option}
```

```python
option_dict
```

```python
df = df.iloc[3::,:]
```

```python
df.columns = ['Date'] + list(colnames)
sum( [ sum (  df.iloc[::,i].isnull()  ) for i in range(116)] )
#checks for missing value
```

```python
df.head()
```

```python
category = [ 'Imported', 'Light Trucks', 'Domestic',
            'Cars', 'Mini', 'Light Vehicles', 'Medium Duty Trucks', 'Hybrid/Electric/Fuel Cell Vehicles' ]
brand = [ i for i in option if i not in category ]
```

```python
brand_pos = [option_dict[i][0] for i in brand]
```

```python
brand_pos1 = np.append(np.array(brand_pos[1::])-1 , 112)
brand_loc = [[brand_pos[i], j] for i,j in enumerate(brand_pos1)]
```

```python
brand_loc,brand
```

```python
new_df = pd.DataFrame(df['Date'])
```

```python
for i,j in enumerate(brand):
    new_df[j] = np.array([ sum(df.iloc[k,option_dict[j]]) for k in range(229)])
```

```python
for i,j in enumerate(category):
    new_df[category[i]] = np.array([ sum(df.iloc[k, option_dict[j]]) for k in range(229) ])
```
