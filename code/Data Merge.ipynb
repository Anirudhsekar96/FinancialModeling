```python
import numpy as np 
import pandas as pd
```

```python
"""
        Target is 1 if there is a postive change versus last timestamp, else it is 0
"""
org_saartot_df           = pd.read_excel("../data/Original_Data/saartot-2000-2019.xlsx",sheet_name="Sheet1",header=None)
org_saartot_df.drop([4],axis=1,inplace=True)
org_saartot_df.dropna(subset=[0],inplace=True)
print(len(org_saartot_df))
org_saartot_df = org_saartot_df.rename(index=str,columns={0:"Timestamp",1:"Value",2:"Absolute_change",3:"Percentage_change"})
org_saartot_df['Target'] = org_saartot_df["Absolute_change"].apply(lambda x: 1 if x>0 else 0)
y = org_saartot_df.Target
org_saartot_df.head()
```

```python
# Checking for class imbalance
print("Percentage of labels which are 1: " + str(y.sum()*100/len(y))[:5] + "%")
print("Percentage of labels which are 0: " + str(100 - (y.sum()*100/len(y)))[:5] + "%")
# There is little to no class imbalance
```

```python
org_saartot_df['Timestamp'] = pd.to_datetime(org_saartot_df['Timestamp'])
org_saartot_df['Year'] = org_saartot_df['Timestamp'].dt.year
org_saartot_df['Month'] = org_saartot_df['Timestamp'].dt.month
org_saartot_df.head()
```

```python
cars_df = pd.read_csv("../data/Cars_.csv")
cars_df["Timestamp"] = pd.to_datetime(cars_df['date'])
cars_df["Year"] = cars_df['Timestamp'].dt.year
cars_df["Month"] = cars_df['Timestamp'].dt.month
cars_df.head()
```

```python
# Merging the target saartot index with the sales data that we have
merged_df = pd.merge(cars_df,org_saartot_df,on=["Year","Month"],how="inner")
merged_df["y"] = merged_df["Target"].shift(1)[:-1]
cols_to_be_dropped = ["date","month","Timestamp_x","Timestamp_y","year"]
merged_df.drop(cols_to_be_dropped,axis=1,inplace=True)
merged_df.dropna(inplace=True)
#merged_df.to_csv('../data/Model_data/training_test_data.csv')
merged_df.head()
```

```python
org_ratio_df = pd.read_excel("../data/Original_Data/ratios for automotive sector players.xlsx")

def quart(x):
    """
        Returns the month number of a quarter in a year
    """
    if(x=='Q1'):
        return(3)
    elif(x=='Q2'):
        return(6)
    elif(x=='Q3'):
        return(9)
    elif(x=='Q4'):
        return(12)
    
    
splits = ["PE_RATIO" ,
          "BEST_EST_PE_CUR_YR",
          "BEST_EST_PE_NXT_YR",
          "PX_TO_BOOK_RATIO",
          "PX_TO_TANG_BOOK_VAL_PER_SH",
          "CURRENT_PX_TO_FREE_CASH_FLOW",
          "PX_TO_SALES_RATIO",
          "EV_TO_T12M_EBITDA",
          "EV_TO_T12M_SALES",
          "EV_TO_MKT_CAP",
          "EV_TO_T12M_CASH_FLOW",
          "EV_TO_T12M_NET_INCOME",
          "EQY_DVD_YLD_IND",
          "EBIT_YIELD",
          "FREE_CASH_FLOW_YIELD",
          "T12_FCF_YIELD"      
         ]

id_vars =  ["Description", "Ticker","Field Mnemonic","Data State","Field ID"]

val_vars= ['2019 Q1', '2018 Q4', '2018 Q3', '2018 Q2', '2018 Q1', '2017 Q4',
       '2017 Q3', '2017 Q2', '2017 Q1', '2016 Q4', '2016 Q3', '2016 Q2',
       '2016 Q1', '2015 Q4', '2015 Q3', '2015 Q2', '2015 Q1', '2014 Q4',
       '2014 Q3', '2014 Q2', '2014 Q1', '2013 Q4', '2013 Q3', '2013 Q2',
       '2013 Q1', '2012 Q4', '2012 Q3', '2012 Q2', '2012 Q1', '2011 Q4',
       '2011 Q3', '2011 Q2', '2011 Q1', '2010 Q4', '2010 Q3', '2010 Q2',
       '2010 Q1', '2009 Q4', '2009 Q3', '2009 Q2', '2009 Q1', '2008 Q4',
       '2008 Q3', '2008 Q2', '2008 Q1', '2007 Q4', '2007 Q3', '2007 Q2',
       '2007 Q1', '2006 Q4', '2006 Q3', '2006 Q2', '2006 Q1', '2005 Q4',
       '2005 Q3', '2005 Q2', '2005 Q1', '2004 Q4', '2004 Q3', '2004 Q2']

drop_cars = ['    Avtovaz PJSC',
             '    BYD Co Ltd',
             '    Brilliance China Automotive Holdings Ltd',
             '    Chongqing Changan Automobile Co Ltd',
             '    DRB-Hicom Bhd',
             '    Daihatsu Motor Co Ltd',
             '    Dongfeng Motor Group Co Ltd',
             '    FAW CAR Co Ltd',
             '    Ford Otomotiv Sanayi AS',
             '    GAZ PJSC',
             '    Geely Automobile Holdings Ltd',
             '    Great Wall Motor Co Ltd',
             '    Guangzhou Automobile Group Co Ltd',
             '    SAIC Motor Corp Ltd',
             '    Sanyang Motor Co Ltd',
             '    Sollers PJSC',
             '    Ssangyong Motor Co',
             '    Suzuki Motor Corp',
             '    Tata Motors Ltd',
             '    Tianjin Faw Xiali Automobile Co Ltd',
             '    Tofas Turk Otomobil Fabrikasi AS',
             '    Yulon Motor Co Ltd',
             '    Peugeot SA', 
             '    Renault SA'
            ]

PE_RATIO                       = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[0]]
BEST_EST_PE_CUR_YR             = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[1]]
BEST_EST_PE_NXT_YR             = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[2]]
PX_TO_BOOK_RATIO               = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[3]]
PX_TO_TANG_BOOK_VAL_PER_SH     = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[4]]
CURRENT_PX_TO_FREE_CASH_FLOW   = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[5]]
PX_TO_SALES_RATIO              = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[6]]
EV_TO_T12M_EBITDA              = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[7]]
EV_TO_T12M_SALES               = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[8]]
EV_TO_MKT_CAP                  = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[9]]
EV_TO_T12M_CASH_FLOW           = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[10]]
EV_TO_T12M_NET_INCOME          = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[11]]
EQY_DVD_YLD_IND                = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[12]]
EBIT_YIELD                     = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[13]]
FREE_CASH_FLOW_YIELD           = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[14]]
T12_FCF_YIELD                  = org_ratio_df[org_ratio_df["Field Mnemonic"] == splits[15]]

PE_RATIO = pd.melt(PE_RATIO,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
BEST_EST_PE_CUR_YR = pd.melt(BEST_EST_PE_CUR_YR,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
BEST_EST_PE_NXT_YR = pd.melt(BEST_EST_PE_NXT_YR,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
PX_TO_BOOK_RATIO = pd.melt(PX_TO_BOOK_RATIO,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
PX_TO_TANG_BOOK_VAL_PER_SH = pd.melt(PX_TO_TANG_BOOK_VAL_PER_SH,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
CURRENT_PX_TO_FREE_CASH_FLOW = pd.melt(CURRENT_PX_TO_FREE_CASH_FLOW,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
PX_TO_SALES_RATIO = pd.melt(PX_TO_SALES_RATIO,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
EV_TO_T12M_EBITDA = pd.melt(EV_TO_T12M_EBITDA,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
EV_TO_T12M_SALES = pd.melt(EV_TO_T12M_SALES,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
EV_TO_MKT_CAP = pd.melt(EV_TO_MKT_CAP,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
EV_TO_T12M_CASH_FLOW = pd.melt(EV_TO_T12M_CASH_FLOW,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
EV_TO_T12M_NET_INCOME = pd.melt(EV_TO_T12M_NET_INCOME,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
EQY_DVD_YLD_IND = pd.melt(EQY_DVD_YLD_IND,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
EBIT_YIELD = pd.melt(EBIT_YIELD,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
FREE_CASH_FLOW_YIELD = pd.melt(FREE_CASH_FLOW_YIELD,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)
T12_FCF_YIELD = pd.melt(T12_FCF_YIELD,id_vars=id_vars,value_vars=val_vars).drop(["Ticker","Field Mnemonic","Data State","Field ID"],axis=1).pivot(index="variable",columns="Description",values=["value"])["value"].drop(drop_cars,axis=1)

dfs_to_be_merged = [PE_RATIO,
                   BEST_EST_PE_CUR_YR,
                   BEST_EST_PE_NXT_YR,
                   PX_TO_BOOK_RATIO,
                   PX_TO_TANG_BOOK_VAL_PER_SH,
                   CURRENT_PX_TO_FREE_CASH_FLOW,
                   PX_TO_SALES_RATIO,
                   EV_TO_T12M_EBITDA,
                   EV_TO_T12M_SALES,
                   EV_TO_MKT_CAP,
                   EV_TO_T12M_CASH_FLOW,
                   EV_TO_T12M_NET_INCOME,
                   EQY_DVD_YLD_IND,
                   EBIT_YIELD,
                   FREE_CASH_FLOW_YIELD,
                   T12_FCF_YIELD  
                  ]

merged_df_ratios = pd.concat(dfs_to_be_merged,axis=1)
merged_df_ratios = merged_df_ratios.reset_index()

merged_df_ratios["Year"] = merged_df_ratios["variable"].apply(lambda x: x.split()[0])
merged_df_ratios["Quarter"] = merged_df_ratios["variable"].apply(lambda x: x.split()[1])
merged_df_ratios["Month"] = merged_df_ratios["Quarter"].apply(lambda x: quart(x))

merged_df_ratios.head()
```

```python
# Building a naive model

split_val = 50
X_train = merged_df.drop(["y"],axis=1)[split_val:]
y_train = merged_df.y[split_val:]

X_test = merged_df.drop(["y"],axis=1)[:split_val]
y_test = merged_df.y[:split_val]


from sklearn.ensemble import GradientBoostingClassifier

GBC = GradientBoostingClassifier()

GBC.fit(X_train,y_train)
y_pred_proba = GBC.predict_proba(X_test)[:,1]

# Returns the mean accuracy
GBC.score(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve,roc_auc_score
%matplotlib inline

fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
auc = roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="ROC AUC= "+str(auc))
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.legend(loc=4)
plt.show()
```
